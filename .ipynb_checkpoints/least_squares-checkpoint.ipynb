```python
from scipy.optimize import least_squares
import numpy as np
import matplotlib.pyplot as plt
import ares
from average_dTb import average_dTb
import os
```

```python
def dTb(theta, z):
    m_chi = theta[0]
    
    directory = "average_dTb/m_chi"+"{:.2f}".format(m_chi)
    if os.path.exists(directory):
        if np.size(os.listdir(directory)) < 150:
            more_random_v_stream = 150 - np.size(os.listdir(directory))
            print("{} more_random_v_streams will be generated for m_chi = {} GeV...".format(more_random_v_stream, m_chi))
        else:
            more_random_v_stream = 0
            z_array, dTb_averaged = np.load(directory+'_averaged.npy')
            print("Adequate files already exist for m_chi = {} GeV. Existing averaged dTb and z loaded.".format(m_chi))
            print("---"*15)
    else:
        more_random_v_stream = 150
        print("{} more random v_streams will be generated for m_chi = {} GeV...".format(more_random_v_stream, m_chi))
    
    if more_random_v_stream:
        z_array, dTb_averaged, m_chi = average_dTb(m_chi=m_chi, more_random_v_stream=more_random_v_stream, mpi=1, verbose=False)
    
    dTb = np.interp(z, z_array, dTb_averaged)
    return dTb

def fun(theta):
    residual = dTb(theta, z_sample) - dTb_sample
    return residual
```

```python
z_sample = np.arange(10, 300, 5)

theta_true = [0.15]
dTb_accurate = dTb(theta_true, z_sample)

noise = 3
dTb_sample = dTb_accurate + noise * np.random.normal(size = z_sample.shape[0])

theta_guess = [0.1]
res = least_squares(fun, theta_guess, diff_step=0.1, bounds=(0, 10), xtol=1e-3)
```

```python
print(res.success)
print(res.status)
print('---'*15)
print('real value =', theta_true)
print('initial guess =', theta_guess)
print('fit =', res.x)

sim = ares.simulations.Global21cm(radiative_transfer=False, verbose=False)
sim.run()

plt.figure(dpi=120)
plt.plot(sim.history['z'], sim.history['dTb'], label = 'no DM heating', color='k', linestyle='--')
plt.plot(z_sample, dTb_accurate, label = r'$m_{\chi, \rm real}$'+' = {} GeV'.format(theta_true[0]))
plt.scatter(z_sample, dTb_sample, label=r'sample, $\sigma_{\rm noise}$'+' = {} mK'.format(noise), s=8)
plt.plot(z_sample, dTb(res.x, z_sample), label = r'$m_{\chi, \rm fit}$'+' = {:.2f} GeV'.format(res.x[0]), linestyle=':', c='r')
plt.xlim(0,300)
# plt.ylim(-60,0)
plt.xlabel(r"$z$")
plt.ylabel(r"$\delta T_b \rm\ [mK]$")
plt.legend()
plt.title(r"infer $m_\chi$ via observing $\delta T_b$")
plt.show()
```
